```python
# Step 7: Change the format of "ORDER_CREATION_DATE" to datetime[64] with the format "%Y%m%d"
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format="%Y%m%d")
df
```

```python
# Step 8: Change the format of "REQUESTED_DELIVERY_DATE" to datetime[64] with the format "%Y%m%d"
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format="%Y%m%d")
df
```

```python
# Step 9: Sanity check - Check how many records have an order date greater than the delivery date
sanity_check = df[df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']]
print("Number of records with order date greater than delivery date:", len(sanity_check))
```

```python
# Step 10: Drop the column where order date is greater than delivery date
df.drop(df[df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']].index, inplace=True)
df
```

```python
# Step 11: Check the number of records where the "ORDER_AMOUNT" field has "-"
count_minus = len(df[df['ORDER_AMOUNT'].str.contains("-")])
print("Number of records with '-' in the ORDER_AMOUNT field:", count_minus)
```

```python
# Step 12: Replace "-" with "" in the "ORDER_AMOUNT" field
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].str.replace("-", "")
```

```python
# Step 13: Check the number of records where the "ORDER_AMOUNT" field has ","
count_comma = len(df[df['ORDER_AMOUNT'].str.contains(",")])
print("Number of records with ',' in the ORDER_AMOUNT field:", count_comma)
```

```python
# Step 14: Replace "," with "." in the "ORDER_AMOUNT" field
df['ORDER_AMOUNT'] = df['ORDER_AMOUNT'].str.replace(",", ".")
df
```

```python
# Step 15: Count the number of records where the order date and delivery date are the same
count_same_dates = len(df[df['ORDER_CREATION_DATE'].dt.date == df['REQUESTED_DELIVERY_DATE'].dt.date])
print("Number of records with the same order and delivery date:", count_same_dates)
```

```python
# Step 16: Count the number of records for each currency type using the "ORDER_CURRENCY" field
#currency_counts = df['ORDER_CURRENCY'].value_counts()
print("Currency counts:")
df['ORDER_CURRENCY'].value_counts()
#print("Currency counts:")
#print(currency_counts)
```

```python
# Step 17: Create a new column "amount_in_usd" and convert non-USD currencies to USD
conversion_rates = {
    'USD': 1.0,
    'EUR': 1.08,
    'AUD': 0.66,
    'CAD': 0.74,
    'GBP': 1.24,
    'MYR': 0.22,
    'PLN': 0.24,
    'AED': 0.27,
    'HKD': 0.13,
    'CHF': 1.11,
    'RON': 0.22,
    'SGD': 0.74,
    'CZK': 0.046,
    'HU1': 0.0028,
    'NZD': 0.61,
    'BHD': 2.65,
    'SAR': 0.27,
    'QAR': 0.27,
    'KWD': 3.25,
    'SEK': 0.094
} 
df['ORDER_AMOUNT']=df['ORDER_AMOUNT'].apply(lambda x:float(x))
df['amount_in_usd']=df.apply(lambda x:x['ORDER_AMOUNT']*conversion_rates.get(x['ORDER_CURRENCY']),axis=1)
df
```

```python
# Step 18: Check for values "0" in the "amount_in_usd" column
count_zero_amounts = len(df[df['amount_in_usd'] == 0])
print("Number of records with amount zero in USD:", count_zero_amounts)
```

```python
# Step 19: Create a new column "unique_cust_id" by adding 'CUSTOMER_NUMBER' and 'COMPANY_CODE'
df['unique_cust_id'] = df['CUSTOMER_NUMBER'].astype(str)+df['COMPANY_CODE'].astype(str)
df
```

```python
from sklearn.metrics import mean_squared_error, r2_score
#TESTING ACCURACY
# Make predictions using the trained models on the test dataset
predictions_linear = linear_regression.predict(X_test) # Make predictions using the Linear Regression model
# predictions_svm = svr.predict(X_test)
predictions_dt = decision_tree.predict(X_test)# Make predictions using the Decision Tree Regression model
predictions_rf = random_forest.predict(X_test)# Make predictions using the Random Forest Regression model
predictions_adaboost = adaboost.predict(X_test) # Make predictions using the AdaBoost Regression model
predictions_xgboost = xgboost.predict(X_test) # Make predictions using the XGBoost Regression model


# Calculate MSE
mse_linear = mean_squared_error(y_test, predictions_linear) # Calculate Mean Squared Error for Linear Regression
# mse_svm = mean_squared_error(y_test, predictions_svm)
mse_dt = mean_squared_error(y_test, predictions_dt) # Calculate Mean Squared Error for Decision Tree Regression
mse_rf = mean_squared_error(y_test, predictions_rf) # Calculate Mean Squared Error for Random Forest Regression
mse_adaboost = mean_squared_error(y_test, predictions_adaboost) # Calculate Mean Squared Error for AdaBoost Regression
mse_xgboost = mean_squared_error(y_test, predictions_xgboost)  # Calculate Mean Squared Error for XGBoost Regression

# Calculate RMSE
rmse_linear = np.sqrt(mse_linear)  # Calculate Root Mean Squared Error for Linear Regression
# rmse_svm = np.sqrt(mse_svm)
rmse_dt = np.sqrt(mse_dt)# Calculate Root Mean Squared Error for Decision Tree Regression
rmse_rf = np.sqrt(mse_rf) # Calculate Root Mean Squared Error for Random Forest Regression
rmse_adaboost = np.sqrt(mse_adaboost) # Calculate Root Mean Squared Error for AdaBoost Regression
rmse_xgboost = np.sqrt(mse_xgboost) # Calculate Root Mean Squared Error for XGBoost Regression

# Calculate R-Squared
r2_linear = r2_score(y_test, predictions_linear) # Calculate R-squared for Linear Regression
# r2_svm = r2_score(y_test, predictions_svm)
r2_dt = r2_score(y_test, predictions_dt)# Calculate R-squared for Decision Tree Regression
r2_rf = r2_score(y_test, predictions_rf) # Calculate R-squared for Random Forest Regression
r2_adaboost = r2_score(y_test, predictions_adaboost) # Calculate R-squared for AdaBoost Regression
r2_xgboost = r2_score(y_test, predictions_xgboost) # Calculate R-squared for XGBoost Regression

print("Linear Regression:")
print("MSE:", mse_linear)
print("RMSE:", rmse_linear)
print("R-Squared:", r2_linear)
print()

# print("Support Vector Machine:")
# print("MSE:", mse_svm)
# print("RMSE:", rmse_svm)
# print("R-Squared:", r2_svm)
#print()

print("Decision Tree:")
print("MSE:", mse_dt)
print("RMSE:", rmse_dt)
print("R-Squared:", r2_dt)
print()

print("Random Forest:")
print("MSE:", mse_rf)
print("RMSE:", rmse_rf)
print("R-Squared:", r2_rf)
print()

print("AdaBoost:")
print("MSE:", mse_adaboost)
print("RMSE:", rmse_adaboost)
print("R-Squared:", r2_adaboost)
print()

print("XGBoost:")
print("MSE:", mse_xgboost)
print("RMSE:", rmse_xgboost)
print("R-Squared:", r2_xgboost)
print()
```

```python
r2_values = {
    'Linear Regression': r2_linear,
    # 'Support Vector Machine': r2_svm,
    'Decision Tree': r2_dt,
    'Random Forest': r2_rf,
    'AdaBoost': r2_adaboost,
    'XGBoost': r2_xgboost
}

# Print the R-Squared values for each model
for model, r2 in r2_values.items():
    print(model, "R-Squared:", r2)

# Find the model with the highest R-Squared value
best_model = max(r2_values, key=r2_values.get)

# Print the best model
print("Best Model based on R-Squared:", best_model)
```

```python
# Create a dictionary to store the MSE values for each model
mse_values = {
    'Linear Regression': mse_linear,
    # 'Support Vector Machine': mse_svm,
    'Decision Tree': mse_dt,
    'Random Forest': mse_rf,
    'AdaBoost': mse_adaboost,
    'XGBoost': mse_xgboost
}

# Calculate the average MSE across all models
average_mse = {model: np.mean(mse_values[model]) for model in mse_values}

# Find the model with the lowest average MSE
best_model = min(average_mse, key=average_mse.get)

# Print the best model
print("Best Model based on MSE:", best_model)
```
